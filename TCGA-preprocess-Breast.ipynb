```python
PatientSC= pd.read_table('/Users/dilrajkaur/Desktop/TCGA-survival/SKCM/SKCM-patient.txt')
```

```python
PatientSC=PatientSC.drop(PatientSC.index[0])
PatientSC=PatientSC.drop(PatientSC.index[0])
PatientSC = PatientSC.drop_duplicates(['bcr_patient_barcode'],keep='first')
PatientSC.to_csv('/Users/dilrajkaur/Desktop/TCGA-survival/SKCM/Heliyon/SKCM-patient-full.csv',index=False,header=True)
```

```python
PatientSC.replace({'[Not Applicable]':'[NA]','[Not Available]':'[NA]','[Discrepancy]':'[NA]','[Completed]':'[NA]'},inplace=True)
##Select the desired columns for analysis
PatientSCF=PatientSC
#PatientSCF = PatientSC.filter(['bcr_patient_barcode','gender','race','ethnicity','anatomic_neoplasm_subdivision','ajcc_staging_edition','ajcc_tumor_pathologic_pt','ajcc_nodes_pathologic_pn','ajcc_metastasis_pathologic_pm','ajcc_pathologic_tumor_stage','vital_status','last_contact_days_to','death_days_to','age_at_diagnosis'], axis=1)
PatientSCF[['last_contact_days_to','death_days_to','age_at_diagnosis']] = PatientSCF[['last_contact_days_to','death_days_to','age_at_diagnosis']].replace('[NA]',0)
PatientSCF[['vital_status']] = PatientSCF[['vital_status']].replace(['Dead','Alive'], [1,0])
PatientSCF['last_contact_days_to'] = pd.to_numeric(PatientSCF['last_contact_days_to'])
PatientSCF['death_days_to'] = pd.to_numeric(PatientSCF['death_days_to'])
PatientSCF['age_at_diagnosis'] = pd.to_numeric(PatientSCF['age_at_diagnosis'])
PatientSCF['OS.time'] = PatientSCF.apply(lambda row: row.last_contact_days_to + row.death_days_to, axis=1)
```

```python
PatientSCF.to_csv('/Users/dilrajkaur/Desktop/TCGA-survival/SKCM/Heliyon/SKCM-patientFinal.csv',index=False,header=True)
LUSC_full=pd.merge(PatientSCF, File2, on='bcr_patient_barcode',how='inner')
LUSC_full = LUSC_full.drop(LUSC_full[LUSC_full['age_at_diagnosis']<=0].index)
LUSC_full = LUSC_full.drop(LUSC_full[LUSC_full['OS.time']<0].index)
LUSC_full.to_csv('/Users/dilrajkaur/Desktop/TCGA-survival/SKCM/Heliyon/SKCM-final.csv',index=False,header=True)
```
